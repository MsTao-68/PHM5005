```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/TracyTao/Desktop/Singapore/NUS/PHM5005/ML/PHM5005/AML/Version4 Apr9/AML (2).csv')
df.shape
```

```python
y = df['SUBTYPES'] # <--- Target (Case/Control)
x = df.drop(['SUBTYPES', 'ID_REF'], axis = 1) # <--- Features (Genomic/Clinical)
print(x.shape, y.shape)
```

```python
x_log_transformed = np.log1p(x) 
```

```python
from scipy.stats import kruskal

# Perform Kruskal-Wallis Test
kruskal_result = kruskal(*[x_log_transformed.iloc[:, i] for i in range(x_log_transformed.shape[1])], y)
print("Kruskal-Wallis Test Result:")
print(kruskal_result)
```

```python
from scipy.stats import kruskal

# Empty list to store significant features
significant_features = []

# Perform Kruskal-Wallis Test for each feature
for feature in x.columns:
    kruskal_result = kruskal(x[feature], y)
    if kruskal_result.pvalue < 0.05:
        significant_features.append(feature)

# Print significant features
print("Significant Features (p-value < 0.05):")
print(significant_features)
```
